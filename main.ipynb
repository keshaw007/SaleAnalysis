# Sales Data

```python
import pandas as pd
import os
```

**Mergin 12 months of sales data in a single .csv file**

```python
df = pd.read_csv("SalesAnalysis/Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('SalesAnalysis/Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("SalesAnalysis/Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data,df])

all_months_data.to_csv("all_data.csv", index=False)
```

**Read in updated dataframes**

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

# Augment data with additional columns

**Cleaning rows with NaN's**

```python
all_data = all_data.dropna(how = 'all')
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int64')
all_data.head()

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

**Adding sales column**

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head(10)
```

# What was the best month for sale?

```python
all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13,1)
results = all_data.groupby('Month').sum()

plt.bar(months, results['Sales'])
plt.ylabel('Sales in USD *10^6')
plt.xlabel('Month')
plt.show()
```

# Which city had the highest number of sales?

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x : x.split(',')[1])
all_data.drop(columns = 'City', inplace = True)
all_data.head()
```

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x : x.split(',')[1])
cities = all_data['City'].unique()
results = all_data.groupby('City').sum()

plt.bar(cities, results['Sales'])

plt.xticks(cities, rotation='vertical', size=8)
plt.xlabel('cityName')
plt.ylabel('Sales in USD')
plt.show()
```

# Which products are sold the most togather?


```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped']= df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count:
    print(key+","+value)
```
